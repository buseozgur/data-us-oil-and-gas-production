## Yearly Gas production

```python
yearly_gas_df = gas_df.groupby(gas_df["Month"].dt.year).sum(numeric_only = True)
yearly_gas_df.index.name = "Year"
yearly_gas_df
```


### Check your code

```python
index_year = yearly_gas_df.index[0]
yearly_gas_shape = yearly_gas_df.shape
us_total = yearly_gas_df.iloc[0,0]
```

```python
from nbresult import ChallengeResult

result = ChallengeResult('full_gas',
    index_year=index_year,
    yearly_gas_shape=yearly_gas_shape,
    us_total=us_total
)
result.write()

print(result.check())
```

```python
plot = yearly_gas_df.filter(items = ["U.S."]).plot(kind = "bar")
plot.set_xlabel("Year");
```

```python
# drop the years where months are missing (2008, 2018)
yearly_gas_df.index
```

```python
condition = np.logical_and(yearly_gas_df.index > yearly_gas_df.index.min(),yearly_gas_df.index < yearly_gas_df.index.max())
```

```python
filtered_yearly_gas_df = yearly_gas_df[condition]
filtered_yearly_gas_df
```

```python
plot = filtered_yearly_gas_df.filter(items = ["U.S."]).plot(kind = "bar")
plot.set_xlabel("Year");
```

### Check your code

```python
from nbresult import ChallengeResult

result = ChallengeResult('filtered_gas',
    yearly_gas=filtered_yearly_gas_df.shape
)
result.write()

print(result.check())
```

## Comparing with Crude Oil Production

```python
file = "data/U.S._crude_oil_production.csv"
oil_df = pd.read_csv(file, decimal=",", parse_dates = ["Month"])
oil_df.head(3)
```

```python
oil_df.shape
```

```python
oil_df.info()
```

---

## Yearly Oil production

```python
yearly_oil_df = oil_df.groupby(oil_df["Month"].dt.year).sum(numeric_only = True)
yearly_oil_df.index.name = "Year"
yearly_oil_df
```

```python
yearly_oil_df.columns = yearly_oil_df.columns.str.strip()
yearly_oil_df.columns
```

```python

plot = yearly_oil_df.filter(items=['U.S. Crude Oil']).plot(kind='bar');
```

```python
yearly_oil_df.index
```

```python
condition = np.logical_and(yearly_oil_df.index > yearly_oil_df.index.min(),yearly_oil_df.index < yearly_oil_df.index.max())
```

```python
filtered_yearly_oil_df = yearly_oil_df[condition]
filtered_yearly_oil_df
```

### Check your code

```python
from nbresult import ChallengeResult

result = ChallengeResult('oil',
    filtered_oil_shape=filtered_yearly_oil_df.shape,
    filtered_oil_index_year=filtered_yearly_oil_df.index[0],
    us_total=filtered_yearly_oil_df.iloc[0,0]
)
result.write()

print(result.check())
```

## Merging Oil and Gas Production

```python
total_gas = filtered_yearly_gas_df[["U.S."]].rename(columns = {"U.S.": "Gas"})
total_gas
```

```python
total_oil = filtered_yearly_oil_df[["U.S. Crude Oil"]].rename(columns = {"U.S. Crude Oil": "Crude Oil"})
total_oil
```

```python
merged_df = pd.concat([total_gas, total_oil], axis=1)
merged_df
```

```python
plot = merged_df.plot(kind="bar")
plot.set_xlabel("Year")
plot.legend(['Gas (Millions of Cubic feet)', 'Crude Oil (Thousands of barrels)']);
```
