## Yearly Gas production

```python
yearly_gas_df = gas_df.groupby(gas_df["Month"].dt.year).sum(numeric_only = True)
yearly_gas_df
```


### Check your code

```python
index_year = yearly_gas_df.index[0]
yearly_gas_shape = yearly_gas_df.shape
us_total = yearly_gas_df.iloc[0,0]
```
